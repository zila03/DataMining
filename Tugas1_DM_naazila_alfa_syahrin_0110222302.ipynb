```python
import pandas as pd
import numpy as np

df = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')

print("Importing Data :")
print(df.to_string()) 
```

```python
grouped_df = df.groupby('Gender')
berat_stats = grouped_df['Weight'].agg(['count', 'mean', 'max', 'min'])
tinggi_stats = grouped_df['Height'].agg(['count', 'mean', 'max', 'min'])

print("Statistik Berat Badan:")
print(berat_stats)

print("\nStatistik Tinggi Badan:")
print(tinggi_stats)
```

```python
df['Gender'] = df['Gender'].replace({'Female':0, 'Male':1})
df['Gender'] = df['Gender'].fillna(0)

print(df)
```

```python
import matplotlib.pyplot as plt

df['BMI'] = df['Weight'] / ((df['Height'] / 100) ** 2)

df.groupby('Gender')['BMI'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Rata-rata BMI')
plt.title('Rata-rata BMI Berdasarkan Gender')
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()
```
